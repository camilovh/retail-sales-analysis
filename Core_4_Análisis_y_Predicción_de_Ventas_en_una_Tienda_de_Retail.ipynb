```python
#total de ventas por categoría de producto
sales_by_category = df.groupby('Product Category')['Total Amount'].sum().reset_index()
display(sales_by_category)
```

```python
#promedio de ventas diarias por categoría de producto.
daily_avg_sales_by_category = df.groupby(['Date', 'Product Category'])['Total Amount'].mean().reset_index()
display(daily_avg_sales_by_category)
```

```python
#Identifica las categorías de productos con mayores y menores ventas.
highest_sales_category = sales_by_category.sort_values(by='Total Amount', ascending=False).iloc[0]
lowest_sales_category = sales_by_category.sort_values(by='Total Amount', ascending=True).iloc[0]

print("Categoría con mayores ventas:")
display(highest_sales_category)

print("\nCategoría con menores ventas:")
display(lowest_sales_category)
```

```python
#ventas para la categoría 'Electronics'
electronics_sales = df[df['Product Category'] == 'Electronics']
display(electronics_sales.head())
```

```python
# 1. Calcular el 'Total Amount' esperado (Multiplicación)
df['Calculated Total Amount'] = df['Quantity'] * df['Price per Unit']

# 2. Calcular la diferencia entre el 'Total Amount' original y el calculado (Resta)
df['Amount Difference'] = df['Total Amount'] - df['Calculated Total Amount']

# 3. Calcular el 'Price per Item Check' para verificar (División)
# Manejo de división por cero si 'Quantity' puede ser 0
df['Price per Item Check'] = df.apply(lambda row: row['Total Amount'] / row['Quantity'] if row['Quantity'] != 0 else 0, axis=1)

# 4. Calcular la proporción de cada venta sobre el total global de ventas (División)
total_global_sales = df['Total Amount'].sum()
df['Sales Proportion'] = df['Total Amount'] / total_global_sales

# 5. Mostrar las nuevas columnas y una vista de las primeras filas
display(df[['Quantity', 'Price per Unit', 'Total Amount', 'Calculated Total Amount', 'Amount Difference', 'Price per Item Check', 'Sales Proportion']].head())
```
